## Exporatory Analysis

```python
df.columns
```

```python
## Select M1 data supply only
df_m1 = df[["Series Description", "M1; Not seasonally adjusted"]]
```

```python
## Set index t0 years
df_m1.set_index("Series Description", inplace = True)
```

```python
df_m1
```

```python
## Select 2021-01 - 2022-12
df_m1 = pd.DataFrame(df_m1.loc['2021-01': "2022-12", "M1; Not seasonally adjusted"])

## Rename columns
df_m1.rename(columns={"Series Description":"Years", "M1; Not seasonally adjusted": "M1 Supply"}, inplace = True)
df_m1.index.names = ["Years"]
```

```python
df_m1
```

```python
## Change datatypes to integers
df_m1["M1 Supply"] = df_m1["M1 Supply"].astype("float")
df_m1
```
